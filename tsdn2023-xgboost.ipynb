```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/attack-simulation-flow/attack_simulation_flow.csv", low_memory=False)
# alert_df = pd.read_csv("/kaggle/input/attack-simulation-lab/attack-simulation-alert.csv", low_memory=False)
```

```python
df['src_ip'].unique()
```

```python
df['labels'] = np.where((df['src_ip'] == '10.20.30.103') | (df['dst_ip'] == '10.20.30.103'), 1, 0)
```

```python
benign_df = pd.read_csv("/kaggle/input/benign-nfstream/benign_nfstream.csv")
benign_df['labels'] = 0
```

```python
df = pd.concat([df, benign_df], ignore_index=True)
```

```python
# pd.set_option("max_rows", 1000)
pd.set_option('display.max_rows', None)
df.describe().T
```

```python
df.info()
```

```python
not_features = [
    "id",
    "expiration_id",
    "src_ip",
    "src_mac",
    "src_oui",
    "src_port",
    "dst_ip",
    "dst_mac",
    "dst_oui",
    "dst_port",
    "protocol",
    "ip_version",
    "vlan_id",
    "tunnel_id",
    "application_name",
    "application_category_name",
    "application_is_guessed",
    "application_confidence",
    "requested_server_name",
    "client_fingerprint",
    "server_fingerprint",
    "user_agent",
    "content_type",
    "labels",
]

features = [c for c in df.columns if c not in not_features]

X = df[features]
y = df['labels']
```

```python
import warnings
import optuna
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier

warnings.simplefilter(action='ignore', category=FutureWarning)

def objective(trial):
    params = {
        'tree_method': 'hist',
        'device': 'cuda',
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.5),
        'subsample': trial.suggest_float('subsample', 0.5, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 30),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'alpha': trial.suggest_float('reg_alpha', 0, 1),
        'lambda': trial.suggest_float('reg_lambda', 0, 1),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 10),
        'max_delta_step': trial.suggest_float('max_delta_step', 0, 10),
    }

    xgb = XGBClassifier(**params)
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    # Use F1 score, precision, and recall as the evaluation metrics
    f1_scorer = make_scorer(score_func=f1_score, greater_is_better=True)
    precision_scorer = make_scorer(score_func=precision_score, greater_is_better=True)
    recall_scorer = make_scorer(score_func=recall_score, greater_is_better=True)

    # Perform cross-validation
    f1_scores = cross_val_score(xgb, X, y, cv=cv, scoring=f1_scorer)
    precision_scores = cross_val_score(xgb, X, y, cv=cv, scoring=precision_scorer)
    recall_scores = cross_val_score(xgb, X, y, cv=cv, scoring=recall_scorer)

    # Compute and print the mean scores
    mean_f1 = f1_scores.mean()
    mean_precision = precision_scores.mean()
    mean_recall = recall_scores.mean()

    print(f'Mean F1 Score: {mean_f1}')
    print(f'Mean Precision: {mean_precision}')
    print(f'Mean Recall: {mean_recall}')

    # Return the average F1 score as the optimization target
    return mean_f1

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

best_params = study.best_params
best_score = study.best_value

print("Best Hyperparameters:", best_params)
print("Best F1 Score:", best_score)
```

```python
from datetime import date

best_xgb = XGBClassifier(**best_params)
best_xgb.fit(X, y)
best_xgb.save_model(f"xgb_{date.today()}_{best_score:.6f}.json")
```
